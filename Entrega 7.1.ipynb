# Hands-on 7: Passo 2 Prática 02: Geração de variáveis aleatórias com uma PDF específica. 

```matlab




clc;clear all; close all;
%% Parâmetros
lbd_1 = 2;                                                 % taxa em segundos/chamadas
lbd_2 = 5;                                                 % taxa em segundos/chamadas
T=0.1;                                                     % Taxa de amostragem
x=0:T:10;                                                 % Eixo x      
% Distribuição 

DistExp_1=exppdf(x,1/lbd_1);                               % Distribuição exponecial  
DistExp_2=exppdf(x,1/lbd_2);                               % Distribuição exponecial

%deveiramos supor 3min como sendo 180s? ou supor os lambdas em minutos e não segundos?

% Estimação da probabilidade
t1= max(find(x>=3));                                     % Acha os índices no qual o valor de y associado ao x seja maior que mu+38 e toma o maior desses índices
t2= min(find(x>=3));                                     % Acha os índices no qual o valor de y associado ao x seja maior que mu+38 e toma o menor desses índices
indices = t2:t1;                                           % Encontra o local que se deseja estimar toamdno a faixa com os indices de interesse.

prob1=sum(DistExp_1(indices))*T*100;                       % Probabilidade de um evento ocorrer no intervalo
prob2=sum(DistExp_2(indices))*T*100;                       % Probabilidade de um evento ocorrer no intervalo

subplot(211)
plot(x,DistExp_1);
title([ 'Probabilidade de =  ' num2str(prob1)  ]);      % Mostra valor verdadeiro de prob1
hold on;
area(x(indices),DistExp_1(indices));                      % Probabilidade 
axis([2.5 10 0 0.01]) ;

subplot(212)
plot(x,DistExp_2); 
title([ 'Probabilidade de =  ' num2str(prob2)  ]);      % Mostra valor verdadeiro de prob1
hold on;
area(x(indices),DistExp_2(indices));                      % Probabilidade 
axis([2.5 10 0 0.000002]) ;

disp([' Probabilidade pela integral da fórmula da PDF = ' num2str(prob1) ' %']);
disp([' Probabilidade pela integral da fórmula da PDF = ' num2str(prob2) ' %']);
```

## DISTRIBUIÇÃO GAUSSIANA

```matlab
clc; clear all;close all;
%% Parâmetros da distribuição
vtMu = [0 -10 10];              % Valores de média da Gaussiana
vtVar = [1 5 10];               % Valores de variância da Gaussiana
x = -20:0.1:20;
sigma = sqrt(vtVar(1));
% Variando a média e plotando os gráficos
fig = figure;
chLegend = [];
for ik=vtMu
    mu = ik;
    vtPDF=normpdf(x,mu,sigma);  % Função que retorna a PDF Gaussiana 
    vtCDF=normcdf(x,mu,sigma);  % Função que retorna a CDF Gaussiana
    subplot(2,2,1);
    plot(x,vtPDF);
    hold all;
    subplot(2,2,2);
    plot(x,vtCDF);
    hold all;
    %
    chLegend = [chLegend; {['Média =  ' num2str(mu )]}];
end
subplot(2,2,1);
legend(chLegend);
subplot(2,2,2);
legend(chLegend);
% Variando a variância e plotando gráficos
mu = vtMu(1);
chLegend = [];
for ik=vtVar
    sigma = sqrt(ik);
    vtPDF=normpdf(x,mu,sigma);
    vtCDF=normcdf(x,mu,sigma);
    subplot(2,2,3);
    plot(x,vtPDF);
    hold all;
    subplot(2,2,4);
    plot(x,vtCDF);
    hold all;
    %
    chLegend = [chLegend; {['\sigma =  ' num2str(sigma)]}];
end
subplot(2,2,3);
legend(chLegend);
subplot(2,2,4);
legend(chLegend);
fig.PaperUnits = 'inches';
fig.PaperPosition = [0 0 12 6];
```

**Para o relatório:** A vida efetiva de um componente de turbina de avião é de 5000h e desvio padrão de 40h. A distribuição da vida efetiva é bastante próxima da distribuição Normal. Faça um código que calcule e mostre graficamente a probabilidade de que uma determinada amostra tenha vida efetiva maior que 5038h. 

```matlab
clc;clear all; close all;
%% Parâmetros
mu = 5000;                                               % Média
sigma = 40.0;                                            % Desvio padrâo
T=0.01;                                                  % Taxa de amostragem
x=0:T:10000;                                             % Eixo x       
% Distribuição
DistNorm=normpdf(x,mu,sigma);                            % Distribuição normal    
% Estimação da probabilidade
t1= max(find(x>=mu+38));                                 % Acha os índices no qual o valor de y associado ao x seja maior que mu+38 e toma o maior desses índices
t2= min(find(x>=mu+38));                                 % Acha os índices no qual o valor de y associado ao x seja maior que mu+38 e toma o menor desses índices
indices = t2:t1;                                         % Encontra o local que se deseja estimar toamdno a faixa com os indices de interesse.
prob1=sum(DistNorm(indices))*T*100;                      % Probabilidade de um evento ocorrer no intervalo
plot(x,DistNorm);   
axis([4850 5150 0 0.01]) ;
title([ 'Probabilidade de =  ' num2str(prob1)  ]);       % Mostra valor verdadeiro de prob1
hold on;                                                 % Não sei o que isso faz.
area(x(indices),DistNorm(indices));                      % Probabilidade (pinta aparte desejada)

disp([' Probabilidade pela integral da fórmula da PDF = ' num2str(prob1) ' %']);
% Outra maneira de calcular
% syms xs fs;
% fs = 1/(sqrt(2*pi*sigma^2))*exp(-(xs-mu)^2/(2*sigma^2));
% prob2 = eval(int(fs,0,sigma))*100;
% disp([' Probabilidade pela área da PDF = ' num2str(prob2) ' %']);
```

**Para o relatório:** A duração de chamadas em uma rede telefônica é uma VA com distribuição exponencial de taxa λ, em segundo/chamadas. Duas operadoras mediram: λ1=2 e λ2=5.
```matlab
clc;clear all; close all;
%% Parâmetros
lbd_1 = 2;                                                 % taxa em segundos/chamadas
lbd_2 = 5;                                                 % taxa em segundos/chamadas
T=0.1;                                                     % Taxa de amostragem
x=0:T:10;                                                 % Eixo x      
% Distribuição 

DistExp_1=exppdf(x,1/lbd_1);                               % Distribuição exponecial  
DistExp_2=exppdf(x,1/lbd_2);                               % Distribuição exponecial

%deveiramos supor 3min como sendo 180s? ou supor os lambdas em minutos e não segundos?

% Estimação da probabilidade
t1= max(find(x>=3));                                     % Acha os índices no qual o valor de y associado ao x seja maior que mu+38 e toma o maior desses índices
t2= min(find(x>=3));                                     % Acha os índices no qual o valor de y associado ao x seja maior que mu+38 e toma o menor desses índices
indices = t2:t1;                                           % Encontra o local que se deseja estimar toamdno a faixa com os indices de interesse.

prob1=sum(DistExp_1(indices))*T*100;                       % Probabilidade de um evento ocorrer no intervalo
prob2=sum(DistExp_2(indices))*T*100;                       % Probabilidade de um evento ocorrer no intervalo

subplot(211)
plot(x,DistExp_1);
title([ 'Probabilidade de \lambda_1 =  ' num2str(prob1),'%'  ]);      % Mostra valor verdadeiro de prob1
hold on;
area(x(indices),DistExp_1(indices));                      % Probabilidade 
axis([2.5 10 0 0.01]) ;

subplot(212)
plot(x,DistExp_2); 
title([ 'Probabilidade de \lambda_2 =  ' num2str(prob2),'%'  ]);      % Mostra valor verdadeiro de prob1
hold on;
area(x(indices),DistExp_2(indices));                      % Probabilidade 
axis([2.5 10 0 0.000002]) ;

disp([' Probabilidade pela integral da fórmula da PDF para \lambda_1 = ' num2str(prob1) ' %']);
disp([' Probabilidade pela integral da fórmula da PDF para \lambda_2 = ' num2str(prob2) ' %']);
```
