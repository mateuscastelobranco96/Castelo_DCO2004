# Hands on 8: Passo 2 Prática 02: Ruídos em comunicações digitais. 

```matlab
clc;clear all; close all;
%% Parâmetros
EbN0_dB_1 = 0;                                          % Eb/N0 de entrada
EbN0_dB_2 = 5;
EbN0_dB_3 = 20;
Ns = 10^5;                                             % Número de símbolos simulados
%% Sinal M-PSK gerado pela função pskmod
M_2 = 2;                                                % Número de símbolo da modulação
M_8 = 8;
bits2 = randi([0 M_2-1],Ns,1);
bits8 = randi([0 M_8-1],Ns,1);

txSig_2 = pskmod(bits2,M_2);                            % Modulação com M=2
txSig_8 = pskmod(bits8,M_8);                            % Modulação com M=8

rxSig_2_1 = awgn(txSig_2,EbN0_dB_1);                    % Modulação com M=2 e Eb/No=0
rxSig_2_2 = awgn(txSig_2,EbN0_dB_2);                    % Modulação com M=2 e Eb/No=5
rxSig_2_3 = awgn(txSig_2,EbN0_dB_3);                    % Modulação com M=2 e Eb/No=20

rxSig_8_1 = awgn(txSig_8,EbN0_dB_1);                    % Modulação com M=8 e Eb/No=0
rxSig_8_2 = awgn(txSig_8,EbN0_dB_2);                    % Modulação com M=8 e Eb/No=5
rxSig_8_3 = awgn(txSig_8,EbN0_dB_3);                    % Modulação com M=8 e Eb/No=20

%% Gráficos M=2
fig = figure(1);
subplot(2,3,1);
plot(rxSig_2_1(find(real(rxSig_2_1)>0)),'ro');
hold on;
plot(rxSig_2_1(find(real(rxSig_2_1)<=0)),'bs')
title('M=2 Eb/No=0');
legend('Bit 1 transmitido','Bit 0 transmitido');
minAx1 = min([real(rxSig_2_1)' imag(rxSig_2_1)']);
maxAx1 = max([real(rxSig_2_1)' imag(rxSig_2_1)']);
% axis([minAx1 maxAx1 minAx1 maxAx1]);

subplot(2,3,2);
plot(rxSig_2_2(find(real(rxSig_2_2)>0)),'ro');
hold on;
plot(rxSig_2_2(find(real(rxSig_2_2)<=0)),'bs')
title('M=2 Eb/No=5');
legend('Bit 1 transmitido','Bit 0 transmitido');
minAx2 = min([real(rxSig_2_2)' imag(rxSig_2_2)']);
maxAx2 = max([real(rxSig_2_2)' imag(rxSig_2_2)']);
% axis([minAx2 maxAx2 minAx2 maxAx2]);

subplot(2,3,3);
plot(rxSig_2_3(find(real(rxSig_2_3)>0)),'ro');
hold on;
plot(rxSig_2_3(find(real(rxSig_2_3)<=0)),'bs')
title('M=2 Eb/No=20');
legend('Bit 1 transmitido','Bit 0 transmitido');
minAx3 = min([real(rxSig_2_3)' imag(rxSig_2_3)']);
maxAx3 = max([real(rxSig_2_3)' imag(rxSig_2_3)']);
axis([minAx3 maxAx3 minAx3 maxAx3]);
fig.PaperUnits = 'inches';
fig.PaperPosition = [0 0 12 6];

%% Gráficos M=8
%fig = figure(2);
subplot(2,3,4);
erro=0.5;

plot(rxSig_8_1(find(abs(rxSig_8_1-1+i*0)<erro)),'ro');
hold on;

plot(rxSig_8_1(find(abs(rxSig_8_1-0.707+i*0.707)<erro)),'bs')
hold on;
plot(rxSig_8_1(find(abs(rxSig_8_1-0.707+i*0.707)<erro)),'ro')
hold on;

plot(rxSig_8_1(find(abs(rxSig_8_1+1+i*0)<erro)),'ro');
hold on;
plot(rxSig_8_1(find(abs(rxSig_8_1+1+i*0)<erro)),'bs');
hold on;

plot(rxSig_8_1(find(abs(rxSig_8_1+0.707-i*0.707)<erro)),'bs')
hold on;
plot(rxSig_8_1(find(abs(rxSig_8_1+0.707-i*0.707)<erro)),'ro')
hold on;

plot(rxSig_8_1(find(abs(rxSig_8_1+0+i*1)<erro)),'ro');
hold on;
plot(rxSig_8_1(find(abs(rxSig_8_1+0+i*1)<erro)),'bs');
hold on;

plot(rxSig_8_1(find(abs(rxSig_8_1-0.707-i*0.707)<erro)),'bs')
hold on;

plot(rxSig_8_1(find(abs(rxSig_8_1+0-i*1)<erro)),'ro');
hold on;
plot(rxSig_8_1(find(abs(rxSig_8_1+0-i*1)<erro)),'bs');
hold on;

plot(rxSig_8_1(find(abs(rxSig_8_1+0.707+i*0.707)<erro)),'bs')
hold on;
plot(rxSig_8_1(find(abs(rxSig_8_1+0.707+i*0.707)<erro)),'ro')

title('M=8 Eb/No=0');
legend('Bit 1 transmitido','Bit 0 transmitido');
minAx4 = min([real(rxSig_8_1)' imag(rxSig_8_1)']);
maxAx4 = max([real(rxSig_8_1)' imag(rxSig_8_1)']);
% axis([minAx4 maxAx4 minAx4 maxAx4]);

subplot(2,3,5);

plot(rxSig_8_2(find(abs(rxSig_8_2-1+i*0)<erro)),'ro');
hold on;

plot(rxSig_8_2(find(abs(rxSig_8_2-0.707+i*0.707)<erro)),'bs')
hold on;
plot(rxSig_8_2(find(abs(rxSig_8_2-0.707+i*0.707)<erro)),'ro')
hold on;

plot(rxSig_8_2(find(abs(rxSig_8_2+1+i*0)<erro)),'ro');
hold on;
plot(rxSig_8_2(find(abs(rxSig_8_2+1+i*0)<erro)),'bs');
hold on;

plot(rxSig_8_2(find(abs(rxSig_8_2+0.707-i*0.707)<erro)),'bs')
hold on;
plot(rxSig_8_2(find(abs(rxSig_8_2+0.707-i*0.707)<erro)),'ro')
hold on;

plot(rxSig_8_2(find(abs(rxSig_8_2+0+i*1)<erro)),'ro');
hold on;
plot(rxSig_8_2(find(abs(rxSig_8_2+0+i*1)<erro)),'bs');
hold on;

plot(rxSig_8_2(find(abs(rxSig_8_2-0.707-i*0.707)<erro)),'bs')
hold on;

plot(rxSig_8_2(find(abs(rxSig_8_2+0-i*1)<erro)),'ro');
hold on;
plot(rxSig_8_2(find(abs(rxSig_8_2+0-i*1)<erro)),'bs');
hold on;

plot(rxSig_8_2(find(abs(rxSig_8_2+0.707+i*0.707)<erro)),'bs')
hold on;
plot(rxSig_8_2(find(abs(rxSig_8_2+0.707+i*0.707)<erro)),'ro')

title('M=8 Eb/No=5');
legend('Bit 1 transmitido','Bit 0 transmitido');
minAx5 = min([real(rxSig_8_2)' imag(rxSig_8_2)']);
maxAx5 = max([real(rxSig_8_2)' imag(rxSig_8_2)']);
axis([minAx5 maxAx5 minAx5 maxAx5]);


subplot(2,3,6);

%111
plot(rxSig_8_3(find(abs(rxSig_8_3-1+i*0)<erro)),'ro');
hold on;
%110
plot(rxSig_8_3(find(abs(rxSig_8_3-0.707-i*0.707)<erro)),'bs')
hold on;
plot(rxSig_8_3(find(abs(rxSig_8_3-0.707-i*0.707)<erro)),'ro')
hold on;
%010
plot(rxSig_8_3(find(abs(rxSig_8_3+0-i*1)<erro)),'ro');
hold on;
plot(rxSig_8_3(find(abs(rxSig_8_3+0+i*1)<erro)),'bs');
hold on;
%011
plot(rxSig_8_3(find(abs(rxSig_8_3+0.707-i*0.707)<erro)),'bs')
hold on;
plot(rxSig_8_3(find(abs(rxSig_8_3+0.707-i*0.707)<erro)),'ro')
hold on;
%001
plot(rxSig_8_3(find(abs(rxSig_8_3+1+i*0)<erro)),'ro');
hold on;
plot(rxSig_8_3(find(abs(rxSig_8_3+1+i*0)<erro)),'bs');
hold on;
%000
plot(rxSig_8_3(find(abs(rxSig_8_3+0.707+i*0.707)<erro)),'bs')
hold on;
%100
plot(rxSig_8_3(find(abs(rxSig_8_3+0+i*1)<erro)),'ro');
hold on;
plot(rxSig_8_3(find(abs(rxSig_8_3+0+i*1)<erro)),'bs');
hold on;
%101
plot(rxSig_8_3(find(abs(rxSig_8_3-0.707+i*0.707)<erro)),'bs')
hold on;
plot(rxSig_8_3(find(abs(rxSig_8_3-0.707+i*0.707)<erro)),'ro')

title('M=8 Eb/No=20');
legend('Bit 1 transmitido','Bit 0 transmitido');
minAx6 = min([real(rxSig_8_3)' imag(rxSig_8_3)']);
maxAx6 = max([real(rxSig_8_3)' imag(rxSig_8_3)']);
axis([-2 2 -2 2]);
fig.PaperUnits = 'inches';
fig.PaperPosition = [0 0 12 6];
```

Podemos observar pelos diagramas que quanto maior a razão Eb/No mais precisoos bits se distribuem no diagrama de constelação, equanto que quanto  menor essa figura de mérito mais espalhado (e consequentemente) sobrepostos o bits se encontram, gerando erros.

Ao comparar as duas modulações PSK com M's diferentes (2 e 8) e mesmo Eb/No podemos destacar a difrença de níveis que estes assumem. Em ambos os casos há sobreposição de bits, porém na modulação 8-PSK eles estão mais organizaos e concentrados.
