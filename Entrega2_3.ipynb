```python
import numpy as np           #importando as bibliotecas necessrias:
import matplotlib.pyplot as plt 
import scipy.io.wavfile as wv 
import os

                             # Parmetros da onda:
tf = 1                       # Tempo de durao da nota
fc = 512                     # Frequncia da nota D
fs = 100*fc                  # Frequencia de amostragem da nota. 
t =np.arange(0,tf+1/fs,1/fs) # Vetor tempo. Para cada elemento do vetor t, haver um elemento em y correspondente.
A = 1                        # Amplitude do sinal

do=A*np.cos(2*np.pi*fc*t)     # Sinal senoidal
re=A*np.cos((9/8)*2*np.pi*fc*t) 
mi=A*np.cos((5/4)*2*np.pi*fc*t) 
fa=A*np.cos((4/3)*2*np.pi*fc*t) 
sol=A*np.cos((3/2)*2*np.pi*fc*t) 
la=A*np.cos((5/3)*2*np.pi*fc*t) 
si=A*np.cos((15/8)*2*np.pi*fc*t) 

plt.figure(1,figsize=[10,7]) # cria instncia da figura para poder alterar seu tamanho
plt.plot(t,do,t,re,t,mi,t,fa,t,sol,t,la,t,si)                # Visualizar o sinal gerado  
plt.legend(['do','re','mi','fa','sol','la','si'])
plt.axis([0,0.002,-1.2,1.2])      # Zoom para melhor visualizao  
plt.show() 
```
